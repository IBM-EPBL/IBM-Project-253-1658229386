```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
data = pd.read_csv('indian_liver_patient.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data[data['Albumin_and_Globulin_Ratio'].isna()]
```

```python
data.mode()
```

```python
#Replace nan's with the mode of the series
data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mode()[0], inplace = True)
```

```python
data_one = pd.get_dummies(data["Gender"])
data_two = pd.concat((data_one, data), axis=1)
data_two = data_two.drop(["Gender"],axis = 1)
data_two = data_two.drop(["Female"],axis = 1)
data = data_two.rename(columns = {"Male":"Gender"})
data
#1 - No disease
#2 - Yes disease
#Gender:
#0 - Female
#1 - Male
```

```python
scores={}
x = data.iloc[:,0:-1]
y = data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2) 
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#Other Models


```python
#Logistic regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 2000)
model = lr.fit(x_train,y_train)
scores["Logistic regression"] = model.score(x_test,y_test)
scores["Logistic regression"]
```

```python
#Naive Bayes

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model = nb.fit(x_train, y_train)
scores["Naive Bayes"] = model.score(x_test, y_test)
scores["Naive Bayes"]
```

```python
#Stochastic Gradient Descent

from sklearn.linear_model import SGDClassifier
sgdc = SGDClassifier()
model = sgdc.fit(x_train, y_train)
scores["Stochastic Gradient Descent"] = model.score(x_test, y_test)
scores["Stochastic Gradient Descent"]
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
model = dtc.fit(x_train, y_train)
scores["Decision Tree"] = model.score(x_test, y_test)
scores["Decision Tree"]
```

#SVM, Random Forest, KNN

```python
#K-Nearest Neighbours

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
model = knn.fit(x_train, y_train)
scores["K-Nearest Neighbours"] = model.score(x_test, y_test)
scores["K-Nearest Neighbours"]
```

```python
#Random Forest

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
model = rfc.fit(x_train, y_train)
scores["Random Forest"] = model.score(x_test, y_test)
scores["Random Forest"]
```

```python
#Support Vector Machine

from sklearn.svm import SVC
svm = SVC()
model = svm.fit(x_train, y_train)
scores["Support Vector Machine"] = model.score(x_test, y_test)
scores["Support Vector Machine"]
```

```python
scoresdf = pd.DataFrame(scores.items(), columns=['Model', 'Accuracy'])
scoresdf.sort_values(by="Accuracy", ascending=False, inplace=True)
scoresdf
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
svm = SVC()
RFmodel = RandomForestClassifier()
KNNmodel = KNeighborsClassifier()
```

```python
svm.fit(x_train, y_train)
SVMpred = svm.predict(x_test)
SVMaccuracy = accuracy_score(SVMpred, y_test)

RFmodel.fit(x_train, y_train)
RFpred = RFmodel.predict(x_test)
RFaccuracy = accuracy_score(RFpred, y_test)

KNNmodel.fit(x_train,y_train)
KNNpred = KNNmodel.predict(x_test)
KNNaccuracy = accuracy_score(KNNpred,y_test)


print("SVM : {0} | RF : {1} | KNN : {2}".format(SVMaccuracy.round(2), RFaccuracy.round(2), KNNaccuracy.round(2)))
```

```python
print("SVM")
print(classification_report(y_test,SVMpred))
print("Random Forest")
print(classification_report(y_test,RFpred))
print("KNN")
print(classification_report(y_test,KNNpred))
```

```python
pickle.dump(svm, open('SVM_10thaug.pkl','wb'))
pickle.dump(RFmodel, open('RF_10thaug.pkl','wb'))
pickle.dump(KNNmodel, open('KNN_10thaug.pkl','wb'))
```
